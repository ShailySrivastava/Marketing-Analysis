```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('/Data_set 2 - Copy.csv')
```

```python
df.head()
```

```python
with open('/Data_set 2 - Copy.csv', 'r') as file:
    next(file)
    avenue_counts = {}
    for line in file:
        avenue = line.strip()
        if avenue in avenue_counts:
            avenue_counts[avenue] += 1
        else:
            avenue_counts[avenue] = 1
```

```python
most_frequent_avenue = max(avenue_counts, key=avenue_counts.get)
max_frequency = avenue_counts[most_frequent_avenue]

print(f"The investment avenue with the highest frequency is '{most_frequent_avenue}' with {max_frequency} occurrences.")
```
